## Top Spenders

```python
#Top Spenders
Top_Spenders_df = purchase_data.groupby(["SN"])
Purchase_Count = Top_Spenders_df["Purchase ID"].count()
Total_Purchase_Value = Top_Spenders_df["Price"].sum()
Average_Purchase_Price = Top_Spenders_df["Price"].mean()
Top_Spenders_df = {"Purchase Count":Purchase_Count,
                         "Average Purchase Price":Average_Purchase_Price,
                         "Total Purchase Value":Total_Purchase_Value}
Top_Spenders_df["Average Purchase Price"] = Top_Spenders_df["Average Purchase Price"].apply('${:.2f}'.format)

#Top_Spenders_df["Total Purchase Value"] = Top_Spenders_df["Total Purchase Value"].apply('${:.2f}'.format)

Top_Spenders_sort_df = pd.DataFrame(Top_Spenders_df).sort_values(["Total Purchase Value"],ascending=False)
#Apply sort then formatting 
Top_Spenders_sort_df["Total Purchase Value"] = Top_Spenders_sort_df["Total Purchase Value"].apply('${:.2f}'.format)
Top_Spenders_sort_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Most popular items

purchase_data[["Item ID","Item Name","Price"]]
item_purchase = purchase_data[["Item ID","Item Name","Price"]]
group_by_item = item_purchase.groupby(["Item ID","Item Name"])
Total_Purchase_Value = group_by_item["Price"].sum()
Purchase_Count = group_by_item["Price"].count()
Item_Price = group_by_item["Price"].mean()
most_popular_item_df ={  "Purchase Count":Purchase_Count,
                            "Item Price":Item_Price,
                            "Total Purchase Value" : Total_Purchase_Value }                                        
most_popular_item_df["Item Price"]     = most_popular_item_df["Item Price"].apply('${:.2f}'.format) 
most_popular_item_df["Total Purchase Value"] =most_popular_item_df["Total Purchase Value"].apply('${:.2f}'.format)
pd.DataFrame(most_popular_item_df).sort_values(["Purchase Count"],ascending=False)
```
